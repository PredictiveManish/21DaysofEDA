## Feature Engineering 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tested.csv')
df.head()
```

```python
df['Cabin'].unique()
```

## Pairing- Siblings/Spouse(SibSp) + Parent CHild (parch) + self = Family

```python
df['Family'] = df['SibSp'] + df['Parch'] + 1
df.drop(columns=['SibSp', 'Parch'], inplace=True)

df
```

```python
df['FarePerPerson'] = df['Fare']/df['Family']
df
```

### Missing Value Handling

```python
type(df)
print(df.columns)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna('Unknown', inplace=True)
while 'Unknown' in df['Embarked'].values:
    print(df[df['Embarked'] == 'Unknown'])
    break  # prevent infinite loop
```

```python
df.sample(20)
```

```python
df['FareRange'] = pd.qcut(df['Fare'], 4, labels=['Low', 'Med', 'High', 'Very High'])
```

```python
df.head(5)
```

```python
df = pd.get_dummies(df, columns=['Embarked', 'Pclass'], drop_first=True)
```

```python
df.head()
```

## Mapping Ordinal Features

```python
df['CabinLevel']= df['Cabin'].str[0]
cabins = {'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'T':8, np.nan:0}
df['CabinLevel'] = df['CabinLevel'].map(cabins)
```
