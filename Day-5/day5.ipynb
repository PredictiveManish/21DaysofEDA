# Encoding & Scaling

```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('D:/Challenges/21DaysofEDA/tested.csv')
```

```python
df.head()
```

### 1. Label Encoding (Sex-1-male, 0-Female)

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df.head()
```

### 2. One Hot Encoding

```python
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
df.head()
```

### 3. Feature Scaling(Standardization, normalization, Min-Max Scaling)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

1. Standardization(mean=0, std=1)

```python
df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])
df.head()
```

2. Min Max Scaling (0 to 1) (Normalization)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
df[['Age','Fare']] = mms.fit_transform(df[['Age','Fare']])

df.head()
```
