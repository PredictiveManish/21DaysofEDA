## Day-2- Data visualization with Seaborn & Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.set(style="darkgrid")
```

```python
df = sns.load_dataset("tips")
print(df.head())
```

# 1. Histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(df['total_bill'],kde=False, color='skyblue', bins=20)
plt.title("Histogram of Total Bill")
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.savefig('histogram.png')
plt.show()
```

## 2. Boxplot

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['total_bill'], color='lightgreen')
plt.title('Boxplot of total Bill')
plt.savefig('boxplot.png')
```

## 3. KDE Plot (Kernel Density Function)

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['total_bill'], color='purple', shade=True)
plt.title('KDE Plot of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.savefig('kdeplot.png')
plt.show()
```

## Visulizing Categorical Data

1. Countplot (for category frequency)

```python
plt.figure(figsize=(6,4))
sns.countplot(x='day',data=df,palette='pastel')
plt.title('Count of Records per Day')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.savefig('countplot.jpg')
plt.show()
```

2. Barplot (Category vs Numeric)

```python
plt.figure(figsize=(6,3))
sns.barplot(x='day',y='total_bill',data=df,palette='muted')
plt.title('Average Total Bill per Day')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.show()
plt.savefig("barplot.png")
```
