# Bivariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('tested.csv')
```

```python
df.head()
```

### 1. Scatter Plot - Numerical vs Numerical

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age',y='Fare',data=df,hue='Sex')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.savefig('scatterplt.jpg')
plt.show()
```

### 2. Heatmap (Correlation Matrix) - Multiple Numerical Variables

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True,cmap='coolwarm',fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig('heatmap.png')
plt.show()
```

### 3. Box Plot - Numerical vs Categorical

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Fare', data=df, hue='Survived', palette='Set2', legend=False)
plt.title('Did Paying More Increase Survival Chances?', fontsize=14)
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare Paid')
plt.ylim(0, 300)
plt.tight_layout()
plt.savefig('survival_fare_boxplot.jpg', dpi=300)
plt.show()
```

### 4. Violin Plot - Like Boxplot, but shows KDE as well

```python
plt.figure(figsize=(6,4))
sns.violinplot(x='Survived',y='Fare',data=df,palette='muted',legend=False)
plt.title('Did Paying More Increase Survival Chances?')
plt.savefig('violin.jpg')
plt.show()
```

### Pairplot - Multiple Bivariate Relationships

```python
sns.pairplot(df[['Age', 'Fare', 'Survived', 'Sex']], hue='Sex', palette='husl')
plt.suptitle('Survival Comparisons by Sex', y=1.02)
plt.tight_layout()
plt.savefig('pairplot_survival.jpg', dpi=300)
plt.show()
```
